```python
## create a line plot of the time series data
from pandas import read_csv
from matplotlib import pyplot
import pandas as pd

#Plots number of responses to Moti survey over time

file_path = 'Individual.csv'
# on the same plot, plot the time series data using a line plot where x axis is the time series data period 'M'and y axis is the sum of the variables 'Volunteer' and 'Staff', volunteer represented by a blue line and staff represented by a orange line
series = read_csv(file_path, header=0, parse_dates=['Survey Data'], index_col='Survey Data')
series.index = pd.to_datetime(series.index, format='mixed').to_period('M')
series = series.sort_index()
series = series.groupby('Survey Data')['Volunteer'].sum()
series.plot()
series = read_csv(file_path, header=0, parse_dates=['Survey Data'], index_col='Survey Data')
series.index = pd.to_datetime(series.index, format='mixed').to_period('M')
series = series.sort_index()
series = series.groupby('Survey Data')['Staff'].sum()
series.plot()
#add legend
pyplot.legend()
#hide the x-axis label
pyplot.xlabel('')
#add title
pyplot.title('Volunteer and staff survey responses over time')
#save plot to file
pyplot.savefig('responses_over_time.png')
pyplot.show()


## prepare same plot but with cumulative totals of volunteer and staff
series = read_csv(file_path, header=0, parse_dates=['Survey Data'], index_col='Survey Data')
series.index = pd.to_datetime(series.index, format='mixed').to_period('M')
series = series.sort_index()
series = series.groupby('Survey Data')['Volunteer'].sum()
series = series.cumsum()
series.plot()
series = read_csv(file_path, header=0, parse_dates=['Survey Data'], index_col='Survey Data')
series.index = pd.to_datetime(series.index, format='mixed').to_period('M')
series = series.sort_index()
series = series.groupby('Survey Data')['Staff'].sum()
series = series.cumsum()
series.plot()
#add legend
pyplot.legend()
#hide the x-axis label
pyplot.xlabel('')
#add title
pyplot.title('Cumulative volunteer and staff survey responses over time')
#save plot to file
pyplot.savefig('responses_over_time_cumulative.png')
pyplot.show()

```
